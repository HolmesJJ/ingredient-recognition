```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TRAIN_DATASET = glob.glob("dataset/train/*")
```

```python
counts = []
for category in TRAIN_DATASET:
    counts.append({"name": os.path.basename(category), "count": len(glob.glob(category + "/*"))})
counts = sorted(counts, key=lambda item: item["count"], reverse=True) 
```

```python
df = pd.DataFrame(counts)
df
```

```python
df[df["count"] <= 100]
```

```python
plt.xticks(rotation=60)
plt.plot(df.head(15)["name"], df.head(15)["count"])
```

```python
plt.xticks(rotation=60)
plt.plot(df.tail(15)["name"], df.tail(15)["count"])
```
